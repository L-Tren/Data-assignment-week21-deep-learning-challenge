## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df_cleaned = application_df.drop(columns=['EIN', 'NAME'])

# Display the updated DataFrame
application_df_cleaned.head()
```

```python
# Determine the number of unique values in each column.
unique_values = application_df_cleaned.nunique()

# Display the number of unique values 
print(unique_values)
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
value_counts = application_df_cleaned['APPLICATION_TYPE'].value_counts()
value_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
cutoff_value = 528
application_types_to_replace = value_counts[value_counts < cutoff_value].index.tolist()


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_value_type = application_df['CLASSIFICATION'].value_counts()
classification_value_type
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_value_counts = application_df['CLASSIFICATION'].value_counts()
classification_value_type_cleaned = classification_value_counts[classification_value_counts > 1]
classification_value_type_cleaned

```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = 778
classifications_to_replace = classification_value_counts[classification_value_counts < cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df.dtypes
```

```python
# Identify categorical columns
categorical_columns = application_df.select_dtypes(include=['object']).index.tolist()

# Convert categorical columns to dummy variables
dummy_df = pd.get_dummies(application_df)

# Display the numeric DataFrame
dummy_df

```

```python
# Split our preprocessed data into our features and target arrays
X = dummy_df.drop('IS_SUCCESSFUL', axis= 'columns')
y = dummy_df['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=100, activation="relu", input_dim= X.shape[1]))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=60, activation="relu"))

#Third hidden layer - added to improve accuracy 
nn.add(tf.keras.layers.Dense(units=30, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
